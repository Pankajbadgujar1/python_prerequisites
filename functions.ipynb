```python
# why use function ?
# To reuse the code 

num = 24
if num%2 ==0:
    print(f"The {num} is even number")
else:
    print(f"The {num} is odd number")

```

```python
def even_odd(num):
    if num%2 ==0:
        print(f"The {num} is even number")
    else:
        print(f"The {num} is odd number")


print(even_odd(45))
```

```python
# veriable lenght arguments ( single value )
# Positional And  Keywords arguments

def print_nums(*args):
    for num in args:
        print(num)

print_nums(4,5,6,7,8,8,6,5) #we can add any number of arguments 


```

```python
# key word arguments  ( in the form of {key-word = kwargs} key = value )
def print_nums(**kwargs):
    for key,value in kwargs.items():
        print(f"{key}: {value} ")

print_nums(name="Pankaj", city="bhod", pin = 2344555)
```

```python
def print_nums(*args,**kwargs):
    for num in args:
        print(num)
    
    for key,value in kwargs.items():
        print(f"{key}: {value} ")

print_nums(2,4,5,'pankaj',name="Pankaj", city="bhod", pin = 2344555)
```

```python
def is_strong_password(password):
    if len(password)<8:
        return False
    if not any(char.isdigit() for char in password):
        return False
    if not any(char.islower() for char in password):
        return False
    if not any(char.isupper() for char in password):
        return False
    if not any(char in '!@#$%^&*()-+' for char in password):
        return False
    return True

# calling function

print(is_strong_password("wekPass"))
print(is_strong_password("strongpPass23#$"))
```

```python
def calculate_total_cost(cart):
    total_cost = 0
    for item in cart:
        total_cost+=item['price']* item['quantity']

    return total_cost
    
cart = [ {'name':'Apple', 'price':0.5,'quantity':4},
         {'name':'Banana', 'price':0.4,'quantity':1},
         {'name':'Orange', 'price':0.3,'quantity':8},
         {'name':'Grephs', 'price':0.2,'quantity':5},
       ]


calculate_total_cost(cart)
```
